```python
def cal_order(data):
    data = data.sort_values(['time'], kind='mergesort')
    data['order'] = range(len(data))
    return data

def cal_u_order(data):
    data = data.sort_values(['time'], kind='mergesort')
    data['u_order'] = range(len(data))
    return data

def refine_time(data):
    data = data.sort_values(['time'], kind='mergesort')
    time_seq = data['time'].values
    time_gap = 1
    for i, da in enumerate(time_seq[0:-1]):
        if time_seq[i] == time_seq[i+1] or time_seq[i] > time_seq[i+1]:
            time_seq[i+1] = time_seq[i+1] + time_gap
            time_gap += 1
    data['time'] = time_seq
    return  data
```

```python
import pandas as pd
import urllib.request
import zipfile
import os

# 1. Download MovieLens 100k
url = "https://files.grouplens.org/datasets/movielens/ml-100k.zip"
zip_path = "ml-100k.zip"
extract_path = "ml-100k"

if not os.path.exists(zip_path):
    print("Downloading MovieLens 100k...")
    urllib.request.urlretrieve(url, zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(".")

# 2. Load Item Metadata (u.item)
# Columns: movie_id | movie_title | release_date | video_release_date | IMDb_URL | unknown | Action | Adventure ...
genre_df = pd.read_csv(
    os.path.join(extract_path, "u.genre"), 
    sep='|', encoding='latin-1', header=None,
    names=['item_id', 'title', 'date', 'video_date', 'url'] + genre_names
)
genre_names = list(genre_df['item_id'])

items_df = pd.read_csv(
    os.path.join(extract_path, "u.item"), 
    sep='|', encoding='latin-1', header=None,
    names=['item_id', 'title', 'date', 'video_date', 'url'] + genre_names
)

# 3. Create Rich Text for LRD (Title + Genres)
# LRD works best with "Language Knowledge", so we combine fields.
def create_text(row):
    genres = [g for g in genre_names if row[g] == 1]
    return f"Title: {row['title']}. Genres: {', '.join(genres)}."

items_df['lrd_text'] = items_df.apply(create_text, axis=1)

# 4. Create the Map
item_text_map = items_df.set_index('item_id')['lrd_text'].to_dict()

# Test it on your ID from the snippet
print(f"Item 73: {item_text_map.get(73)}") 
# Expected Output: "Title: Maverick (1994). Genres: Action, Comedy, Western."
```

```python
import pandas as pd
import wikipedia
import time
import os

# --- 1. Load your existing dataframe ---
# Assuming you have the 'items_df' or 'final_df' from the previous step 
# that already has 'movie_title' and 'Genre'.
# If not, reload it:
# items_df = pd.read_csv(...) 

# Let's use the subset logic again to test
#target_subset = items_df.head(5).copy() 
target_subset = items_df.copy() 

# --- 2. Define the Wikipedia Fetcher ---
def get_wiki_description(title, year):
    # Construct a specific search query to avoid ambiguity
    # e.g., "Toy Story 1995 film"
    query = f"{title} {year} film"
    
    try:
        # 1. Search for pages
        results = wikipedia.search(query)
        if not results:
            return "No results found"
        
        # 2. Get the most likely page (usually the first one)
        # We wrap in try/catch because sometimes the first result is a disambiguation page
        try:
            page = wikipedia.page(results[0], auto_suggest=False)
        except wikipedia.DisambiguationError as e:
            # If ambiguous, try the first option suggested
            page = wikipedia.page(e.options[0], auto_suggest=False)
        except wikipedia.PageError:
            return "Page not found"
            
        # 3. Return the summary (first 3 sentences are usually enough for LRD)
        return page.summary.split('\n')[0]  # First paragraph
        
    except Exception as e:
        return f"Error: {e}"

# --- 3. Run the Fetch ---
print("Fetching Wikipedia descriptions...")
descriptions = []

for idx, row in target_subset.iterrows():
    # Extract year for better search precision
    try:
        # Title format: "Toy Story (1995)"
        clean_title = row['movie_title'].split('(')[0].strip()
        year_str = row['movie_title'].split('(')[1].replace(')', '')
    except:
        clean_title = row['movie_title']
        year_str = ""

    print(f"Searching: {clean_title}...")
    desc = get_wiki_description(clean_title, year_str)
    descriptions.append(desc)
    
    # Polite delay
    time.sleep(0)

# --- 4. Attach to DataFrame ---
target_subset['Description'] = descriptions

# For LRD, we combine everything into one 'Knowledge' text
target_subset['LRD_Input_Text'] = (
    "movie_title: " + target_subset['movie_title'] + ". " +
    "Genre: " + target_subset['Genre'] + ". " +
    "Summary: " + target_subset['Description']
)

print("\n--- Result for LRD ---")
print(target_subset[['movie_title', 'LRD_Input_Text']].head().values)
```
