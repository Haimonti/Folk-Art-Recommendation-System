```python
def Concept_Rec(user_preferences,train_copy,test_copy,already_recommended,total_user_preferences): 
    #print("User Selected Preferences are : ",user_preferences)              
    #display_selected_images(folder_path,user_preferences)         
    smlrty=[]
    #smlrty_node=[]    
    
    for i in user_preferences:     
        list={} 
        #print()
        #only_node={}
        A=test_copy[test_copy['nodes_identity']==i].values.tolist()[0]
        A = A[:-1]        
        A = [float(x) for x in A]
        for train in range(0,85):   
            B=train_copy.iloc[train].tolist()
            B=B[:-1]
            B = [float(x) for x in B]
            CS=np.dot(A,B)/(norm(A)*norm(B))
            NI=train_copy.iloc[train]['nodes_identity']
            if NI not in total_user_preferences and NI not in already_recommended:
                list[NI]=CS
        smlrty.append(list)
      
    result = []
    
    for d in smlrty:
        sorted_items = sorted(d.items(), key=lambda x: x[1], reverse=True)
        top_7 = dict(sorted_items[:7])
        result.append(top_7)
    return result       
```

```python
def folk_recommendation(user_preferences):
    UP_animal = [element for element in user_preferences if element in animal_label]
    UP_tree = [element for element in user_preferences if element in tree_label]
    UP_myth = [element for element in user_preferences if element in myth_label]
    N_AMT = [element for element in user_preferences if element not in animal_label and element not in tree_label and element not in myth_label]   

    UP_animal = UP_animal + N_AMT
    UP_tree = UP_tree + N_AMT
    UP_myth = UP_myth + N_AMT
    Number_animal = len(UP_animal)
    Number_tree = len(UP_tree)
    Number_myth = len(UP_myth)

    
    Total = Number_animal + Number_tree +  Number_myth 
    Prop_animal =  Number_animal/Total
    Prop_tree = Number_tree/Total
    Prop_myth = Number_myth/Total

    Recc_animal = round(Prop_animal * 7)
    Recc_tree = round(Prop_tree * 7)
    Recc_myth = round(Prop_myth * 7)

    if (Recc_animal+ Recc_tree +Recc_myth)>7:
        H=max([Recc_animal,Recc_tree,Recc_myth])
        if H== Recc_animal:
            Recc_animal-=1
        elif H==Recc_tree:
            Recc_tree-=1            
        elif H==Recc_myth:
            Recc_myth-=1    
    if (Recc_animal+ Recc_tree +Recc_myth)<7:
        H=max([Recc_animal,Recc_tree,Recc_myth])
        if H== Recc_animal:
            Recc_animal+=1
        elif H==Recc_tree:
            Recc_tree+=1            
        elif H==Recc_myth:
            Recc_myth+=1    

    ##print('Recc_animal',Recc_animal)
    ##print('Recc_tree',Recc_tree )
    ##print('Recc_myth',Recc_myth)
    Animal_recommended = []
    if len(UP_animal)!=0:
        recommended=[]
        result = Concept_Rec(UP_animal,train_animal_copy,test_animal_copy,Animal_recommended,user_preferences)
        merged_dict = {k: v for d in result for k, v in d.items()}
        #only_node = {k: v for d in  only_node for k, v in d.items()}
        
        sorted_top_2 = dict(sorted(merged_dict.items(), key=lambda item: item[1], reverse=True)[:Recc_animal])
        
        for key,value in sorted_top_2.items():
            recommended.append(int(key))
        Animal_recommended=recommended

    if len(UP_tree)!=0:
        recommended=[]
       
        result= Concept_Rec(UP_tree,train_tree_copy,test_tree_copy,Animal_recommended,user_preferences)
        merged_dict = {k: v for d in result for k, v in d.items()}
        sorted_top_2 = dict(sorted(merged_dict.items(), key=lambda item: item[1], reverse=True)[:Recc_tree])
        for key,value in sorted_top_2.items():
            recommended.append(int(key))
            Animal_recommended.append(int(key))


    if len(UP_myth)!=0:
        recommended=[]
       
        result  = Concept_Rec(UP_myth,train_myth_copy,test_myth_copy,Animal_recommended,user_preferences)
        merged_dict = {k: v for d in result for k, v in d.items()}
        sorted_top_2 = dict(sorted(merged_dict.items(), key=lambda item: item[1], reverse=True)[:Recc_myth])
        ##print('sorted for myth',sorted_top_2)    

        for key,value in sorted_top_2.items():
            recommended.append(int(key))
            Animal_recommended.append(int(key))
        #print(recommended)
        ##print("Recommended Paintings from myth are : ")
        ##display_selected_images(folder_path, recommended) 
    return Animal_recommended
```

```python
def data_preparation(user_preferences):
    user_animal_preferences={}
    user_tree_preferences={}
    user_myth_preferences={}
    proportion_list_animal = {}
    proportion_list_tree = {}
    proportion_list_myth = {}

    print("user_preferences ",user_preferences )
    for key,value in user_preferences.items() :
        UP_Animal = [element for element in value if element in animal_label]
        print("value",value)
        print("UP_Animal",UP_Animal)
        UP_Tree = [element for element in value if element in tree_label]
        UP_Myth = [element for element in value if element in myth_label]
        N_Amt = [element for element in value if element not in animal_label and element not in tree_label and element not in myth_label]   

        UP_Animal = UP_Animal + N_Amt
        UP_Tree = UP_Tree + N_Amt
        UP_Myth = UP_Myth + N_Amt

        if len(UP_Animal) !=0:
            user_animal_preferences[key]=sorted(UP_Animal)

        if len(UP_Tree) !=0:
            user_tree_preferences[key]=sorted(UP_Tree)
    
        if len(UP_Myth) !=0:
            user_myth_preferences[key]=sorted(UP_Myth)            
            
        Number_Animal = len(UP_Animal)
        Number_Tree = len(UP_Tree)
        Number_Myth = len(UP_Myth)

        
        Total = Number_Animal + Number_Tree +  Number_Myth 
        Prop_animal =  Number_Animal/Total
        Prop_tree = Number_Tree/Total
        Prop_myth = Number_Myth/Total

        Recc_animal = round(Prop_animal * 7)
        Recc_tree = round(Prop_tree * 7)
        Recc_myth = round(Prop_myth * 7)

        if (Recc_animal+ Recc_tree +Recc_myth)>7:
            H=max([Recc_animal,Recc_tree,Recc_myth])
            if H== Recc_animal:
                Recc_animal-=1
            elif H==Recc_tree:
                Recc_tree-=1            
            elif H==Recc_myth:
                Recc_myth-=1    
        if (Recc_animal+ Recc_tree +Recc_myth)<7:
            H=max([Recc_animal,Recc_tree,Recc_myth])
            if H== Recc_animal:
                Recc_animal+=1
            elif H==Recc_tree:
                Recc_tree+=1            
            elif H==Recc_myth:
                Recc_myth+=1 
        proportion_list_animal[key] = Recc_animal  
        proportion_list_tree[key] = Recc_tree        
        proportion_list_myth[key] = Recc_myth        
        
    return user_animal_preferences,user_tree_preferences,user_myth_preferences,proportion_list_animal,proportion_list_tree,proportion_list_myth    
```

```python
animal_MMGCN_dict=animal_MMGCN.T.to_dict('list')
tree_MMGCN_dict=tree_MMGCN.T.to_dict('list')
myth_MMGCN_dict=myth_MMGCN.T.to_dict('list')
```

```python
def remove_test(reco_dict,sorted_list):
    for key,value in reco_dict.items():
        reco_dict[key] = [i for i in value if i not in sorted_list]
    return reco_dict

animal_MMGCN_dict_wt = remove_test(animal_MMGCN_dict, sorted_list)
tree_MMGCN_dict_wt = remove_test(tree_MMGCN_dict, sorted_list)
myth_MMGCN_dict_wt = remove_test(myth_MMGCN_dict, sorted_list)
```

```python
for key,value in tree_MMGCN_dict_wt.items():
    tree_MMGCN_dict_wt[key] = [i for i in value if i not in animal_MMGCN_dict_wt[key]]
    
```

```python
for key,value in myth_MMGCN_dict_wt.items():
    myth_MMGCN_dict_wt[key] = [i for i in value if (i not in animal_MMGCN_dict_wt[key] and i not in tree_MMGCN_dict_wt[key])]
```

```python
for key,value in animal_MMGCN_dict_wt.items():
    animal_MMGCN_dict_wt[key] = value[0:proportion_list_animal[key]]

for key,value in tree_MMGCN_dict_wt.items():
    tree_MMGCN_dict_wt[key] = value[0:proportion_list_tree[key]]

for key,value in myth_MMGCN_dict_wt.items():
    myth_MMGCN_dict_wt[key] = value[0:proportion_list_myth[key]]
```

```python
Recommended_form_MMGCN = {}
for key in range(0,1000):
    Recommended_form_MMGCN[key] = [item for sublist in [animal_MMGCN_dict_wt[key],tree_MMGCN_dict_wt[key],myth_MMGCN_dict_wt[key]] for item in sublist]    
```

```python
def precision_recall(Relevant_items,recommended_items):
    hit={}
    num_hits=[]
    precision =[]
    recall=[]
    for i in range(0,1000):
        #print(i)
        #print(Relevant_items)
        #print(list(set(Relevant_items[i]) & set(recommended_items[i])))
        hit[i]=list(set(Relevant_items[i]) & set(recommended_items[i]))
        num_hits.append(len(hit[i]))  
        recall.append(len(hit[i])/len(Relevant_items[i]))
        precision.append(len(hit[i])/7)
    R=sum(recall)/2000
    P=sum(precision)/2000
    return R,P                                       
    
```

```python
Precision_MMGCN,R_Recall_MMGCN=precision_recall(relevant_items,Recommended_form_MMGCN)   
print("R_Recall_MMGCN",round(Precision_MMGCN,4))
print("R_Recall_MMGCN",round(R_Recall_MMGCN,4))
```
